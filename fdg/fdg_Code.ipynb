二、正規化

```python
def normalize(data):
    from sklearn.preprocessing import MinMaxScaler
    scaler = MinMaxScaler()
    scaler.fit(data)
    data[data.columns] = scaler.fit_transform(data[data.columns].to_numpy()) 
    return data
```

```python
x_train=normalize(train_info)
x_train
```

```python
from sklearn.preprocessing import MinMaxScaler
train_ans1 = MinMaxScaler().fit(train_ans.values.reshape(-1,1))
train_ans2 = MinMaxScaler().fit_transform(train_ans.values.reshape(-1,1))
train_ans3=np.ravel(train_ans2)
train_ans3
```

```python
predict_data2=normalize(predict_data1)
predict_data2
```

```python
predict_data3=predict_data2.drop(columns=['ID'],axis=1)
predict_data3
```

七、績效評估(Logloss)

```python
def logloss(y_true, y_pred, eps=1e-15):
    y_true = np.array(y_true)
    y_pred = np.array(y_pred)
    assert (len(y_true) and len(y_true) == len(y_pred))
    p = np.clip(y_pred, eps, 1-eps)
    loss = np.sum(- y_true * np.log(p) - (1 - y_true) * np.log(1-p))
    return loss / len(y_true)
```

```python
print('多層感知器的logloss:',logloss(train_ans3,mlp_pred))
```

```python
print('隨機森林的logloss:',logloss(train_ans3,rf_pred))
```

```python
print('KNN的logloss:',logloss(train_ans3,knn_pred))
```

```python
print('支援向量機的logloss:',logloss(train_ans3,svm_pred))
```
